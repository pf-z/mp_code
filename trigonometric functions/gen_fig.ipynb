```python
import numpy as np
import matplotlib.pyplot as plt

# 定义通过单位圆旋转来获取sin和cos的函数
def unit_circle(theta):
    # 初始点在单位圆上的位置
    point = np.array([1, 0])
    
    # 旋转矩阵
    rotation_matrix = np.array([
        [np.cos(theta), -np.sin(theta)],
        [np.sin(theta), np.cos(theta)]
    ])
    
    # 旋转后的点
    rotated_point = np.dot(rotation_matrix, point)
    
    # 返回旋转后的点的 y 和 x 值，分别对应 sin(θ) 和 cos(θ)
    return rotated_point[1], rotated_point[0]

# 角度范围：从 -2π 到 2π，包含多个角度
theta_vals = np.linspace(-2 * np.pi, 2 * np.pi, 400)

# 创建一个图形和多个子图（1*3的布局）
fig, axes = plt.subplots(3, 3, figsize=(18, 12))

# 绘制sin和cos图像
ax1 = axes[0, 0]
sin_vals, cos_vals = np.sin(theta_vals), np.cos(theta_vals)
ax1.plot(theta_vals, sin_vals, label="sin(θ)", color='blue')
ax1.set_title("Sine Function")
ax1.set_xlabel("θ (in terms of π)")
ax1.set_ylabel("sin(θ)")
ax1.grid(True)

ax2 = axes[0, 1]
ax2.plot(theta_vals, cos_vals, label="cos(θ)", color='red')
ax2.set_title("Cosine Function")
ax2.set_xlabel("θ (in terms of π)")
ax2.set_ylabel("cos(θ)")
ax2.grid(True)

# 合并sin和cos
ax3 = axes[0, 2]
ax3.plot(theta_vals, sin_vals, label="sin(θ)", color='blue')
ax3.plot(theta_vals, cos_vals, label="cos(θ)", color='red')
ax3.set_title("Sine & Cosine Functions")
ax3.set_xlabel("θ (in terms of π)")
ax3.set_ylabel("Function Value")
ax3.grid(True)
ax3.legend()

# 绘制tan和cot图像
ax4 = axes[1, 0]
tan_vals = np.tan(theta_vals)
ax4.plot(theta_vals, tan_vals, label="tan(θ)", color='green')
ax4.set_ylim(-10, 10)  # 限制y轴范围，避免正切函数的无穷大
ax4.set_title("Tangent Function")
ax4.set_xlabel("θ (in terms of π)")
ax4.set_ylabel("tan(θ)")
ax4.grid(True)

ax5 = axes[1, 1]
cot_vals = 1 / tan_vals
ax5.plot(theta_vals, cot_vals, label="cot(θ)", color='purple')
ax5.set_ylim(-10, 10)  # 限制y轴范围，避免正切函数的无穷大
ax5.set_title("Cotangent Function")
ax5.set_xlabel("θ (in terms of π)")
ax5.set_ylabel("cot(θ)")
ax5.grid(True)

# 合并tan和cot
ax6 = axes[1, 2]
ax6.plot(theta_vals, tan_vals, label="tan(θ)", color='green')
ax6.plot(theta_vals, cot_vals, label="cot(θ)", color='purple')
ax6.set_ylim(-10, 10)  # 限制y轴范围，避免正切函数的无穷大
ax6.set_title("Tangent & Cotangent Functions")
ax6.set_xlabel("θ (in terms of π)")
ax6.set_ylabel("Function Value")
ax6.grid(True)
ax6.legend()

# 绘制sec和csc图像
ax7 = axes[2, 0]
sec_vals = 1 / cos_vals  # 正割：sec(θ) = 1 / cos(θ)
ax7.plot(theta_vals, sec_vals, label="sec(θ)", color='orange')
ax7.set_ylim(-10, 10)  # 限制y轴范围，避免正割函数的无穷大
ax7.set_title("Secant Function")
ax7.set_xlabel("θ (in terms of π)")
ax7.set_ylabel("sec(θ)")
ax7.grid(True)

ax8 = axes[2, 1]
csc_vals = 1 / sin_vals  # 余割：csc(θ) = 1 / sin(θ)
ax8.plot(theta_vals, csc_vals, label="csc(θ)", color='brown')
ax8.set_ylim(-10, 10)  # 限制y轴范围，避免余割函数的无穷大
ax8.set_title("Cosecant Function")
ax8.set_xlabel("θ (in terms of π)")
ax8.set_ylabel("csc(θ)")
ax8.grid(True)

# 合并sec和csc
ax9 = axes[2, 2]
ax9.plot(theta_vals, sec_vals, label="sec(θ)", color='orange')
ax9.plot(theta_vals, csc_vals, label="csc(θ)", color='brown')
ax9.set_ylim(-10, 10)  # 限制y轴范围，避免正割函数的无穷大
ax9.set_title("Secant & Cosecant Functions")
ax9.set_xlabel("θ (in terms of π)")
ax9.set_ylabel("Function Value")
ax9.grid(True)
ax9.legend()

# 自定义横轴标签，减少标记的数量，每 90度 显示一个标签
def format_tick_labels(axis):
    # 每 90度（π/2）标记一个点
    tick_values = np.linspace(-2 * np.pi, 2 * np.pi, 5)  # 5个标记：-2π, -π, 0, π, 2π
    tick_labels = ["-2π", "-π", "0", "π", "2π"]
    
    axis.set_xticks(tick_values)
    axis.set_xticklabels(tick_labels)

# 应用自定义格式化
for ax in axes.flat:
    format_tick_labels(ax)

# 添加总标题并调整布局
fig.suptitle("Trigonometric Functions Visualized Using Unit Circle", fontsize=16)
plt.tight_layout()

plt.savefig('trigonometric functions/fig.png', dpi=300)

plt.show()
```
