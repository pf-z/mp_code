```python
import numpy as np
import matplotlib.pyplot as plt

# sincos
def unit_circle(theta):
    # 
    point = np.array([1, 0])
    
    # 
    rotation_matrix = np.array([
        [np.cos(theta), -np.sin(theta)],
        [np.sin(theta), np.cos(theta)]
    ])
    
    # 
    rotated_point = np.dot(rotation_matrix, point)
    
    #  y  x  sin()  cos()
    return rotated_point[1], rotated_point[0]

#  -2  2
theta_vals = np.linspace(-2 * np.pi, 2 * np.pi, 400)

# 1*3
fig, axes = plt.subplots(3, 3, figsize=(18, 12))

# sincos
ax1 = axes[0, 0]
sin_vals, cos_vals = np.sin(theta_vals), np.cos(theta_vals)
ax1.plot(theta_vals, sin_vals, label="sin()", color='blue')
ax1.set_title("Sine Function")
ax1.set_xlabel(" (in terms of )")
ax1.set_ylabel("sin()")
ax1.grid(True)

ax2 = axes[0, 1]
ax2.plot(theta_vals, cos_vals, label="cos()", color='red')
ax2.set_title("Cosine Function")
ax2.set_xlabel(" (in terms of )")
ax2.set_ylabel("cos()")
ax2.grid(True)

# sincos
ax3 = axes[0, 2]
ax3.plot(theta_vals, sin_vals, label="sin()", color='blue')
ax3.plot(theta_vals, cos_vals, label="cos()", color='red')
ax3.set_title("Sine & Cosine Functions")
ax3.set_xlabel(" (in terms of )")
ax3.set_ylabel("Function Value")
ax3.grid(True)
ax3.legend()

# tancot
ax4 = axes[1, 0]
tan_vals = np.tan(theta_vals)
ax4.plot(theta_vals, tan_vals, label="tan()", color='green')
ax4.set_ylim(-10, 10)  # y
ax4.set_title("Tangent Function")
ax4.set_xlabel(" (in terms of )")
ax4.set_ylabel("tan()")
ax4.grid(True)

ax5 = axes[1, 1]
cot_vals = 1 / tan_vals
ax5.plot(theta_vals, cot_vals, label="cot()", color='purple')
ax5.set_ylim(-10, 10)  # y
ax5.set_title("Cotangent Function")
ax5.set_xlabel(" (in terms of )")
ax5.set_ylabel("cot()")
ax5.grid(True)

# tancot
ax6 = axes[1, 2]
ax6.plot(theta_vals, tan_vals, label="tan()", color='green')
ax6.plot(theta_vals, cot_vals, label="cot()", color='purple')
ax6.set_ylim(-10, 10)  # y
ax6.set_title("Tangent & Cotangent Functions")
ax6.set_xlabel(" (in terms of )")
ax6.set_ylabel("Function Value")
ax6.grid(True)
ax6.legend()

# seccsc
ax7 = axes[2, 0]
sec_vals = 1 / cos_vals  # sec() = 1 / cos()
ax7.plot(theta_vals, sec_vals, label="sec()", color='orange')
ax7.set_ylim(-10, 10)  # y
ax7.set_title("Secant Function")
ax7.set_xlabel(" (in terms of )")
ax7.set_ylabel("sec()")
ax7.grid(True)

ax8 = axes[2, 1]
csc_vals = 1 / sin_vals  # csc() = 1 / sin()
ax8.plot(theta_vals, csc_vals, label="csc()", color='brown')
ax8.set_ylim(-10, 10)  # y
ax8.set_title("Cosecant Function")
ax8.set_xlabel(" (in terms of )")
ax8.set_ylabel("csc()")
ax8.grid(True)

# seccsc
ax9 = axes[2, 2]
ax9.plot(theta_vals, sec_vals, label="sec()", color='orange')
ax9.plot(theta_vals, csc_vals, label="csc()", color='brown')
ax9.set_ylim(-10, 10)  # y
ax9.set_title("Secant & Cosecant Functions")
ax9.set_xlabel(" (in terms of )")
ax9.set_ylabel("Function Value")
ax9.grid(True)
ax9.legend()

#  90 
def format_tick_labels(axis):
    #  90/2
    tick_values = np.linspace(-2 * np.pi, 2 * np.pi, 5)  # 5-2, -, 0, , 2
    tick_labels = ["-2", "-", "0", "", "2"]
    
    axis.set_xticks(tick_values)
    axis.set_xticklabels(tick_labels)

# 
for ax in axes.flat:
    format_tick_labels(ax)

# 
fig.suptitle("Trigonometric Functions Visualized Using Unit Circle", fontsize=16)
plt.tight_layout()

plt.savefig('trigonometric functions/fig.png', dpi=300)

plt.show()
```
