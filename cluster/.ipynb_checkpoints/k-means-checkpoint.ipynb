```python
# ============================================================================
# Author: pf.zh
# Date: 2024-12-02
# ============================================================================

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# 生成数据点，三个簇，确保簇之间不要离得太远
np.random.seed(42)
X = np.vstack([
    np.random.normal(loc=[5, 5], scale=0.4, size=(100, 2)),  # 第一个簇，中心在 (5, 5)
    np.random.normal(loc=[7, 5], scale=0.6, size=(100, 2)),  # 第二个簇，中心在 (7, 5)
    np.random.normal(loc=[6, 7], scale=0.6, size=(100, 2))   # 第三个簇，中心在 (6, 7)
])

# 真实的簇中心
true_centroids = np.array([[5, 5], [7, 5], [6, 7]])

# 创建绘图
fig, ax = plt.subplots(figsize=(8, 6))  # 创建单个子图

# 使用不同的颜色标记真实数据簇
ax.scatter(X[:100, 0], X[:100, 1], c='blue', s=30, alpha=0.5, label="Cluster 1")
ax.scatter(X[100:200, 0], X[100:200, 1], c='green', s=30, alpha=0.5, label="Cluster 2")
ax.scatter(X[200:, 0], X[200:, 1], c='orange', s=30, alpha=0.5, label="Cluster 3")

# 真实的簇中心（用红色的 `x` 标记）
ax.scatter(true_centroids[:, 0], true_centroids[:, 1], c='red', marker='x', s=100, label="True Centroids")
ax.set_title("True Data & Centroids")
ax.legend(loc='best')

plt.savefig('kmeans-fig1.png', dpi=300)

# 显示图形
plt.show()
```

```python
# 创建绘图
fig, ax = plt.subplots(figsize=(8, 6))  # 创建单个子图

# 使用不同的颜色标记真实数据簇
ax.scatter(X[:, 0], X[:, 1], c='grey', s=30, alpha=0.5, label="Cluster 1")

ax.set_title("True Data")


plt.savefig('kmeans-fig2.png', dpi=300)

# 显示图形
plt.show()
```

```python

# K-means 聚类函数
def kmeans_iterate(X, k=3, max_iter=100):
    # 使用随机数生成簇中心，确保初始化的簇中心不会离得太远
    np.random.seed(4)
    centroids = np.random.randn(k, 2) * 0.5 + np.mean(X, axis=0)  # 初始化较为集中的簇中心
    labels = np.zeros(X.shape[0])

    def update(frame):
        nonlocal centroids, labels

        # 计算每个点到各簇中心的距离，并分配簇标签
        distances = np.linalg.norm(X[:, np.newaxis] - centroids, axis=2)
        labels = np.argmin(distances, axis=1)

        # 更新簇中心
        new_centroids = np.array([X[labels == i].mean(axis=0) for i in range(k)])
        
        # 让更新的变化更平滑
        centroids = centroids + 0.3 * (new_centroids - centroids)  # 控制更新幅度，增加平滑

        # 清空右侧图像并重新绘制
        ax[1].clear()
        # 绘制右侧的聚类过程
        colors = ['green', 'orange', 'blue']
        # 为每个簇分配一个颜色
        for i in range(k):
            ax[1].scatter(X[labels == i, 0], X[labels == i, 1], c=colors[i], s=30, alpha=0.5, label=f"Cluster {i+1}")
        #ax[1].scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=30)
        ax[1].scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=100, label="Centroids")
        ax[1].set_title(f"K-means Iteration {frame + 1}")
        ax[1].legend(loc='best')

    return update
```

```python
# 创建绘图和动画
fig, ax = plt.subplots(1, 2, figsize=(12, 6))  # 创建1行2列的子图

# 左侧图：显示真实数据和簇中心
# 使用不同的颜色标记真实数据簇
ax[0].scatter(X[:, 0], X[:, 1], c='grey', s=30, alpha=0.5, label="Cluster 1")

ax[0].set_title("True Data")

# K-means 迭代动画，frames设置为10帧
update_function = kmeans_iterate(X)
ani = FuncAnimation(fig, update_function, frames=range(10), repeat=False)

# 显示图形
plt.show()

# 保存动图
ani.save('kmeans-fig3.gif', writer='imagemagick', fps=1)
```
