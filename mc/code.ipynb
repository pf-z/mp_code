```python
import matplotlib.pyplot as plt
import numpy as np

# 创建图形和坐标轴
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_aspect('equal')

# 绘制单位圆
theta = np.linspace(0, 2 * np.pi, 1000)
x_circle = np.cos(theta)
y_circle = np.sin(theta)
ax.plot(x_circle, y_circle, 'b', label='Unit Circle')

# 绘制外切正方形
square_x = [-1, 1, 1, -1, -1]
square_y = [-1, -1, 1, 1, -1]
ax.plot(square_x, square_y, 'r--', label='Circumscribed Square')

# 填充单位圆的面积
ax.fill(x_circle, y_circle, 'skyblue', alpha=0.5, edgecolor='none', label='Circle Area')

# 标记面积字母
ax.text(0.5, 0.5, '$\pi$', fontsize=14, color='black')

# 设置坐标轴范围和标签
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.axhline(0, color='grey', linestyle='--')
ax.axvline(0, color='grey', linestyle='--')

# 去掉图形边框
for spine in ax.spines.values():
    spine.set_visible(False)

# 显示图例
ax.legend(loc='upper left', fontsize=10, frameon=False)

# 显示图形
plt.title('Unit Circle with Circumscribed Square')


plt.savefig('./Unit_Circle_with_Square.png', dpi=300, bbox_inches='tight')

plt.show()
```
