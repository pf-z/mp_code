```python
# ============================================================================
# Author: pf.zh
# Date: 2024-11-25
# ============================================================================

import numpy as np
import matplotlib.pyplot as plt

# 随机生成二维空间的点，分布朝右上方倾斜
np.random.seed(42)  # 确保结果可复现
n_points = 100
x = np.random.uniform(0, 10, n_points)
y = x + np.random.normal(0, 2, n_points)  # y与x正相关，加上噪声

# 绘制
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color="blue", alpha=0.7, s=30, edgecolor="black", label="Beans")
plt.title("Scattered beans", fontsize=14, fontweight="normal")
plt.legend(fontsize=12, loc="upper left")

# 去掉坐标轴
plt.xticks([])
plt.yticks([])
plt.box(False)

plt.savefig('output/fig1.png', dpi=300)
plt.show()
```

```python
# 数据中心化
x_centered = x - np.mean(x)
y_centered = y - np.mean(y)

# 绘制图形
plt.figure(figsize=(8, 6))
plt.scatter(x_centered, y_centered, color="blue", alpha=0.7, s=30, edgecolor="black", label="Beans")
plt.title("Scattered Beans in two-dimensional Euclidean space", fontsize=14, fontweight="normal")
plt.legend(fontsize=12, loc="upper left")

# 去掉边框
for spine in plt.gca().spines.values():
    spine.set_visible(False)

# 添加坐标轴名称，放置在对应轴箭头旁边
plt.text(x_centered.max() + 2, -0.5, 'x', fontsize=14, ha='center', va='center')
plt.text(-0.5, y_centered.max() + 2, 'y', fontsize=14, ha='center', va='center')

# 使用 plt.arrow 绘制箭头
plt.arrow(x_centered.min() - 2, 0, x_centered.max() - x_centered.min() + 4, 0, 
          head_width=0.5, head_length=0.3, fc='k', ec='k', length_includes_head=True, clip_on=False)

plt.arrow(0, y_centered.min() - 2, 0, y_centered.max() - y_centered.min() + 4, 
          head_width=0.3, head_length=0.5, fc='k', ec='k', length_includes_head=True, clip_on=False)

# 显示网格
plt.grid(alpha=0.3)
plt.savefig('output/fig2.png', dpi=300)
plt.show()
```

```python

from sklearn.decomposition import PCA

# 使用PCA进行主成分分析
data = np.vstack([x_centered, y_centered]).T # 构造二维数据集
pca = PCA(n_components=2) # 两个主成分
pca.fit(data)

# 获取方差最大的方向和方差最小的方向
principal_axis = pca.components_[0]  # 方差最大方向
secondary_axis = pca.components_[1]  # 方差最小方向

# 绘制图形
plt.figure(figsize=(8, 6))
plt.scatter(x_centered, y_centered, color="blue", alpha=0.7, s=30, edgecolor="black", label="Beans")
plt.title("Two directions in two-dimensional space", fontsize=14, fontweight="normal")

# 去掉边框
for spine in plt.gca().spines.values():
    spine.set_visible(False)

# 添加坐标轴名称，放置在对应轴箭头旁边
plt.text(x_centered.max() + 2, -0.5, 'x', fontsize=14, ha='center', va='center')
plt.text(-0.5, y_centered.max() + 2, 'y', fontsize=14, ha='center', va='center')

# 使用 plt.arrow 绘制箭头
plt.arrow(x_centered.min() - 2, 0, x_centered.max() - x_centered.min() + 4, 0, 
          head_width=0.5, head_length=0.3, fc='k', ec='k', length_includes_head=True, clip_on=False)
plt.arrow(0, y_centered.min() - 2, 0, y_centered.max() - y_centered.min() + 4, 
          head_width=0.3, head_length=0.5, fc='k', ec='k', length_includes_head=True, clip_on=False)

# 绘制方差最大的方向向量（主成分）
plt.arrow(-principal_axis[0] * 8, -principal_axis[1] * 8, principal_axis[0] * 18, principal_axis[1] * 18, head_width=0.3, head_length=0.2, width=0.005, fc='red', ec='red', 
          length_includes_head=True, clip_on=False, label="Direction 1")

# 绘制方差最小的方向向量（次主成分）
plt.arrow(-secondary_axis[0] * 6, -secondary_axis[1] * 6, secondary_axis[0] * 12, secondary_axis[1] * 12, head_width=0.3, head_length=0.2, width=0.005, fc='green', ec='green', 
          length_includes_head=True, clip_on=False, label="Direction 2")

plt.legend(fontsize=12, loc="upper left")
# 显示网格
plt.grid(alpha=0.3)
plt.savefig('output/fig3.png', dpi=300)
plt.show()
```

```python
# 计算每个点在主成分上的投影
projected_data_pc1 = data.dot(principal_axis)  # 投影到第一主成分
projected_data_pc2 = data.dot(secondary_axis)  # 投影到第二主成分

# 计算投影点坐标
# 数据在第一主成分上的投影
reconstructed_pc1 = np.outer(projected_data_pc1, principal_axis)
# 数据在第二主成分上的投影
reconstructed_pc2 = np.outer(projected_data_pc2, secondary_axis)

# 绘制图形
plt.figure(figsize=(8, 6))
plt.scatter(x_centered, y_centered, color="#f4f4f4", alpha=0.7, s=30, edgecolor="black", label="Beans")

flag = True
# 绘制投影点和垂线
for i in range(n_points):
    # 第一主成分的垂线
    plt.plot([x_centered[i], reconstructed_pc1[i, 0]], [y_centered[i], reconstructed_pc1[i, 1]], color="#DF6159", linestyle="--", alpha=0.5)
    # 第二主成分的垂线
    plt.plot([x_centered[i], reconstructed_pc2[i, 0]], [y_centered[i], reconstructed_pc2[i, 1]], color="#61DE7D", linestyle="--", alpha=0.5)
    
    # 绘制在主成分上的投影点
    plt.scatter(reconstructed_pc1[i, 0], reconstructed_pc1[i, 1], color='red', marker='o', label=f"{'Projection point in direction 1' if flag else ''}")  # 第一主成分上的投影点
    plt.scatter(reconstructed_pc2[i, 0], reconstructed_pc2[i, 1], color='green', marker='*', label=f"{'Projection point in direction 2' if flag else ''}")  # 第二主成分上的投影点
    if flag:
        flag = False
        
# 绘制方差最大的方向向量（主成分）
plt.arrow(-principal_axis[0] * 8, -principal_axis[1] * 8, principal_axis[0] * 18, principal_axis[1] * 18, head_width=0.3, head_length=0.2, width=0.005, fc='red', ec='red', 
          length_includes_head=True, clip_on=False, label="Direction 1")

# 绘制方差最小的方向向量（次主成分）
plt.arrow(-secondary_axis[0] * 6, -secondary_axis[1] * 6, secondary_axis[0] * 12, secondary_axis[1] * 12, head_width=0.3, head_length=0.2, width=0.005, fc='green', ec='green', 
          length_includes_head=True, clip_on=False, label="Direction 2")

# 设置坐标范围
plt.xlim(np.min(x_centered) - 2, np.max(x_centered) + 2)
plt.ylim(np.min(y_centered) - 2, np.max(y_centered) + 2)

# 显示网格
plt.grid(alpha=0.3)

# 标题和标签
plt.title("The projection process and results of beans", fontsize=14, fontweight="normal")
plt.xlabel("x", fontsize=14)
plt.ylabel("y", fontsize=14)

plt.legend(fontsize=12, loc="upper left")
# 显示图形
plt.savefig('output/fig4.png', dpi=300)
plt.show()
```

```python
# 创建两个子图（左图和右图）
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

flag = True
for i in range(n_points):
    ax1.scatter(reconstructed_pc1[i, 0], reconstructed_pc1[i, 1], color='red', marker='o', label=f"{'Projection point in direction 1' if flag else ''}")  # 第一主成分上的投影点
    if flag:
        flag = False
    
# 绘制方差最大的方向向量（第一主成分）
ax1.arrow(-principal_axis[0] * 8, -principal_axis[1] * 8, principal_axis[0] * 18, principal_axis[1] * 18,
          head_width=0.3, head_length=0.2, width=0.005, fc='red', ec='red', length_includes_head=True, clip_on=False, label="Direction 1")

# 设置左图坐标范围，标题与标签
ax1.set_xlim(np.min(x_centered) - 2, np.max(x_centered) + 2)
ax1.set_ylim(np.min(y_centered) - 2, np.max(y_centered) + 2)
ax1.set_title("Projection on First Principal Component", fontsize=14)
ax1.set_xlabel("x", fontsize=14)
ax1.set_ylabel("y", fontsize=14)
ax1.grid(alpha=0.3)
ax1.legend(fontsize=12, loc="upper left")

flag = True
for i in range(n_points):
    ax2.scatter(reconstructed_pc2[i, 0], reconstructed_pc2[i, 1], color='green', marker='*', label=f"{'Projection point in direction 2' if flag else ''}")  # 第二主成分上的投影点
    if flag:
        flag = False

# 绘制方差最小的方向向量（第二主成分）
ax2.arrow(-secondary_axis[0] * 6, -secondary_axis[1] * 6, secondary_axis[0] * 12, secondary_axis[1] * 12,
          head_width=0.3, head_length=0.2, width=0.005, fc='green', ec='green', length_includes_head=True, clip_on=False, label="Direction 2")

# 设置右图坐标范围，标题与标签
ax2.set_xlim(np.min(x_centered) - 2, np.max(x_centered) + 2)
ax2.set_ylim(np.min(y_centered) - 2, np.max(y_centered) + 2)
ax2.set_title("Projection on Second Principal Component", fontsize=14)
ax2.set_xlabel("x", fontsize=14)
ax2.set_ylabel("y", fontsize=14)
ax2.grid(alpha=0.3)
ax2.legend(fontsize=12, loc="upper left")

# 调整布局并显示图形
plt.tight_layout()
plt.savefig('output/fig5.png', dpi=300)
plt.show()
```

```python
# 创建两个子图（左图和右图）
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# 左图：原始空间
ax1.scatter(x_centered, y_centered, color="blue", alpha=0.7, s=30, edgecolor="black", label="Beans")
ax1.set_title("Two dimensional Euclidean space", fontsize=14, fontweight="normal")
ax1.legend(fontsize=12, loc="upper left")
# 添加坐标轴名称，放置在对应轴箭头旁边
ax1.text(x_centered.max() + 2, -0.5, 'x', fontsize=14, ha='center', va='center')
ax1.text(-0.5, y_centered.max() + 2, 'y', fontsize=14, ha='center', va='center')
# 使用 plt.arrow 绘制箭头
ax1.arrow(x_centered.min() - 2, 0, x_centered.max() - x_centered.min() + 4, 0, 
          head_width=0.5, head_length=0.3, fc='k', ec='k', length_includes_head=True, clip_on=False)
ax1.arrow(0, y_centered.min() - 2, 0, y_centered.max() - y_centered.min() + 4, 
          head_width=0.3, head_length=0.5, fc='k', ec='k', length_includes_head=True, clip_on=False)
# 显示网格
ax1.grid(alpha=0.3)


# 创建网格，设置间距为1
grid_range = np.arange(-8, 8, 1)  # 网格的范围从-8到8，步长为1
x_grid, y_grid = np.meshgrid(grid_range, grid_range)  # 创建网格
# 将网格坐标转换为一列向量
grid_points = np.vstack([x_grid.ravel(), y_grid.ravel()]).T
# 计算主成分方向的旋转矩阵
# 旋转矩阵将坐标系旋转到主成分方向
rotation_matrix = np.array([[principal_axis[0], secondary_axis[0]], 
                            [principal_axis[1], secondary_axis[1]]])
# 对网格进行旋转
rotated_grid_points = grid_points.dot(rotation_matrix)


# 右图：主成分空间
ax2.scatter(x_centered, y_centered, color="#f4f4f4", alpha=0.7, s=30, edgecolor="black", label="Beans")
# 绘制方差最大的方向向量（主成分）
ax2.arrow(-principal_axis[0] * 8, -principal_axis[1] * 8, principal_axis[0] * 18, principal_axis[1] * 18,
          head_width=0.3, head_length=0.2, width=0.005, fc='red', ec='red', length_includes_head=True, clip_on=False,
          label="First principal component")
# 绘制方差最小的方向向量（次主成分）
ax2.arrow(-secondary_axis[0] * 6, -secondary_axis[1] * 6, secondary_axis[0] * 12, secondary_axis[1] * 12,
          head_width=0.3, head_length=0.2, width=0.005, fc='green', ec='green', length_includes_head=True,
          clip_on=False, label="Second principal component")
# 绘制旋转后的网格背景：按行绘制网格线
for i in range(len(grid_range)):
    ax2.plot(rotated_grid_points[i::len(grid_range), 0], rotated_grid_points[i::len(grid_range), 1], color="lightblue", alpha=0.5)
# 绘制旋转后的网格背景：按列绘制网格线
for j in range(len(grid_range)):
    ax2.plot(rotated_grid_points[j*len(grid_range):(j+1)*len(grid_range), 0], 
             rotated_grid_points[j*len(grid_range):(j+1)*len(grid_range), 1], color="lightblue", alpha=0.5)

# 设置坐标范围
ax2.set_xlim(np.min(x_centered) - 2, np.max(x_centered) + 2)
ax2.set_ylim(np.min(y_centered) - 2, np.max(y_centered) + 2)
# 标题和标签
ax2.set_title("The space spaned by two principal components", fontsize=14, fontweight="normal")
ax2.set_xlabel("x", fontsize=14)
ax2.set_ylabel("y", fontsize=14)
# 显示图形
ax2.legend(fontsize=12, loc="upper left")

# 移除边框和刻度
for ax in [ax1, ax2]:
    ax.spines['top'].set_visible(False)   # 去掉上边框
    ax.spines['right'].set_visible(False)  # 去掉右边框
    ax.spines['left'].set_visible(False)   # 去掉左边框
    ax.spines['bottom'].set_visible(False) # 去掉下边框
    ax.tick_params(which='both', bottom=False, top=False, left=False, right=False)  # 去掉刻度

plt.tight_layout()
plt.savefig('output/fig6.png', dpi=300)
plt.show()
```

```python
# 投影数据到主成分方向
projected_data = data.dot(pca.components_.T)  # 使用 dot product 投影到主成分方向

# 可视化原始数据和投影后的数据
plt.figure(figsize=(10, 6))

# 原始数据可视化，坐标轴也在数据中心
plt.subplot(1, 2, 1)
plt.scatter(x_centered, y_centered, label='Original Data', color='blue', alpha=0.6, s=30)
plt.title('Original Data (Centered)')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=1)  # 添加水平坐标轴
plt.axvline(0, color='black', linewidth=1)  # 添加垂直坐标轴
plt.grid(True)
plt.axis('equal')

# 投影后的数据可视化
plt.subplot(1, 2, 2)
plt.scatter(projected_data[:, 0], projected_data[:, 1], label='Projected Data', color='red', alpha=0.6, s=30)
plt.title('Projected Data (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.axhline(0, color='black', linewidth=1)  # 添加水平坐标轴
plt.axvline(0, color='black', linewidth=1)  # 添加垂直坐标轴
plt.grid(True)
plt.axis('equal')

plt.tight_layout()
plt.savefig('output/fig7.png', dpi=300)
plt.show()
```
