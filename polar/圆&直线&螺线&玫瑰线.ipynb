```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Times New Roman, Songti SC'  # 替换为你选择的字体

# 设置图形
fig, axs = plt.subplots(2, 2, subplot_kw={'projection': 'polar'}, figsize=(10, 8))
fig.suptitle('极坐标下的曲线图示', fontsize=16)

# 1. 圆：r = a
theta = np.linspace(0, 2 * np.pi, 500)
r1 = np.full_like(theta, 2)  # 半径为2
axs[0, 0].plot(theta, r1, label='r = 2')
axs[0, 0].set_title('圆')

# 2. 直线：θ = θ₀（如 θ = π/4）
theta_line = np.full(100, np.pi / 4)
r2 = np.linspace(0, 3, 100)
axs[0, 1].plot(theta_line, r2, label='θ = π/4')
axs[0, 1].set_title('直线')

# 3. 阿基米德螺线：r = aθ
theta3 = np.linspace(0, 4 * np.pi, 500)
a = 0.3
r3 = a * theta3
axs[1, 0].plot(theta3, r3, label='r = 0.3θ')
axs[1, 0].set_title('螺线（阿基米德螺线）')

# 4. 玫瑰线：r = a * cos(kθ)，k = 4（偶数 => 8个花瓣）
k = 4
a = 2
theta4 = np.linspace(0, 2 * np.pi, 1000)
r4 = a * np.cos(k * theta4)
axs[1, 1].plot(theta4, r4, label='r = 2cos(4θ)')
axs[1, 1].set_title('玫瑰线（k=4）')

# 美化
for ax in axs.flat:
    ax.legend(loc='upper right')
    ax.grid(True)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
